```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('master.csv')
df.head()
```

```python
df.query('year>1990 and country in[])
```

```python
df['rate_per_100k']=df['suicides_no']/(df['population']/100000)
```

```python
usa_df=df.loc[df['country']=='United States',]
usa_df.reset_index(drop=True,inplace=True)
```

```python
usa_df.drop(['generation','suicides/100k pop','country-year','HDI for year','gdp_per_capita ($)','rate_per_100l'],axis=1)
```

```python
ysa=usa_df.query('year>1990 and year<1995').groupby(['year','sex','age'])['suicides_no'].agg('sum')
ysa
```

```python
fig, ax = plt.subplots(figsize=(15,7))
usa_df.query('year>1990 and year<1995').groupby(['year','sex','age'])['suicides_no'].agg('sum').unstack().plot(kind='bar',ax=ax)
plt.grid(True)
```

```python
usa_ages=sns.catplot(x='age', y='suicides_no', data=usa_df, palette='hls',kind='bar',
                     order=['5-14 years','15-24 years', '25-34 years','35-54 years','55-74 years',
                            '75+ years']).set_xticklabels(rotation=30)
```

```python
g= sns.catplot(x='sex', y='suicides_no', data=usa_df, palette='hls', kind='bar',order=['female','male'])
```

```python
usa_age=sns.barplot(x='age', y='suicides_no', data= usa_df, palette='BuGn_r',order=['5-14 years','15-24 years', '25-34 years','35-54 years',
                    '55-74 years','75+ years'])
usa_age.set_xticklabels(usa_age.get_xticklabels(),rotation=30)
usa_age.set_xlabel="Age"
usa_age.set_ylabel='Number of Suicides'
usa_age.set_title('Number of Suicides per Age Group in the USA: 1985-2015')
```

```python
usa_age=sns.barplot(x='age', y='rate_per_100k', data= usa_df, palette='winter',order=['5-14 years','15-24 years', '25-34 years','35-54 years',
                    '55-74 years','75+ years'])
usa_age.set_xticklabels(usa_age.get_xticklabels(),rotation=30)
usa_age.set_xlabel="Age"
usa_age.set_ylabel='Number of Suicides per 100k'
usa_age.set_title('Number of Suicides per Age Group in the USA: 1985-2015')
```

```python
sns.set_style("whitegrid")
g=sns.catplot(x="sex",y="rate_per_100k", hue="age", kind="bar", data=usa_df, palette='PRGn',order=['female','male'])
(g.fig.suptitle('Number of suicides per 100k: USA'))
```

```python
sns.set_style("whitegrid")
usa_year_s = sns.barplot(x='year',y='rate_per_100k',data=usa_df)
                
usa_year_s.set_xticklabels(usa_year_s.get_xticklabels(), rotation=90)
usa_year_s.set_xlabel('Year')
usa_year_s.set_ylabel('Number of Suicides per 100k people')
usa_year_s.set_title('Number of Suicides per Year: USA')
```

```python
usa_decade=usa_df.loc[usa_df['year'].isin(['1985','1995','2005','2015'])]
```

```python
sns.set_style('whitegrid')
g=sns.catplot(x='age',y='rate_per_100k',col='sex',hue='year',kind='bar',data=usa_decade, order=['5-14 years','15-24 years','25-34 years','35-54 years',
                      '55-74 years','75+ years']).set_xticklabels(rotation=90)

(g.despine(left=True),g.set_axis_labels("", "Number of Suicides per 100k people"))
```
